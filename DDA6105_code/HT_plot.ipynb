```python
plt.figure(figsize = (16, 8))
for directories in os.listdir(os.getcwd())[1:-3]:
  DQN_arr = []
  for i in range(1,11):
    DQN = np.load(directories + "/test_q_{}.npy".format(i))
    DQN_arr.append(DQN)
  DQN_arr = np.array(DQN_arr)
  DQN_mean = np.mean(DQN_arr, axis = 0)
  #DQN_std = np.std(DQN_arr, axis = 0)
  plt.plot(np.arange(500) + 1, DQN_mean, label = directories)
  #plt.fill_between(np.arange(500) + 1, DQN_mean - DQN_std, DQN_mean + DQN_std)
plt.margins(0)
plt.ylim(0.2, 0.3)
plt.rcParams['axes.linewidth'] = 2.5
plt.title("Environment 2", fontsize = 20, weight = "bold")
plt.xlabel("epoch", fontsize = 20, weight = "bold")
plt.ylabel("Average Predicted Value", fontsize = 20, weight = "bold")
plt.xticks(fontsize = 18, weight = "bold")
plt.yticks(fontsize = 18, weight = "bold")
plt.legend(fontsize = 20, bbox_to_anchor = (1.1, 1.015))
plt.savefig("HT.png", dpi = 800)
```
